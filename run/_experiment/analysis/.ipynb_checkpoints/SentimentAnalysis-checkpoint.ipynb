```python
import json
import matplotlib.pyplot as plt
import pandas as pd

# Corrected JSON data
# Load JSON data
with open('Analysis.json') as f:
    json_data = json.load(f)

# Convert the analysis part of JSON to DataFrame
df = pd.DataFrame(json_data['analysis'])

# Create a scatter plot without displaying the file name for x axis
# We will use index of the dataframe as x-axis
plt.figure(figsize=(10, 5))
plt.scatter(df.index, df['positiveWords'], color='green', label='Positive Words')
plt.scatter(df.index, df['negativeWords'], color='red', label='Negative Words')
plt.scatter(df.index, df['stopWords'], color='black', label='Stop Words')

# Adding title and labels
plt.title('Word Analysis from Texts')
plt.xlabel('Document Index')
plt.ylabel('Word Count')
plt.legend()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# The JSON data provided by the user (with corrected 'TrieSearchParallel' spelling)
data = [
    {"totalFiles": 2, "Algorithm": "AhoCorasickParallel", "Total Time Taken": 10},
    {"totalFiles": 2, "Algorithm": "AhoCorasickSequential", "Total Time Taken": 16},
    {"totalFiles": 2, "Algorithm": "TrieSearchParallel", "Total Time Taken": 20},
    {"totalFiles": 2, "Algorithm": "TrieSearchSequential", "Total Time Taken": 25},
    {"totalFiles": 20, "Algorithm": "AhoCorasickParallel", "Total Time Taken": 100},
    {"totalFiles": 20, "Algorithm": "AhoCorasickSequential", "Total Time Taken": 160},
    {"totalFiles": 20, "Algorithm": "TrieSearchParallel", "Total Time Taken": 200},
    {"totalFiles": 20, "Algorithm": "TrieSearchSequential", "Total Time Taken": 250}
]

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(12, 6))

# Get unique algorithms sorted alphabetically for consistent color assignment
algorithms = sorted(df['Algorithm'].unique())

for algorithm in algorithms:
    subset = df[df['Algorithm'] == algorithm]
    plt.plot(subset['totalFiles'], subset['Total Time Taken'], marker='o', linestyle='-', label=algorithm)

plt.title('Total Time Taken by Algorithm against Total Files')
plt.xlabel('Total Files')
plt.ylabel('Total Time Taken')
plt.legend()
plt.grid(True)
plt.show()

```
